```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pbp=r'C:\Users\Anyone\Desktop\Python\Pandas\Jupyterlab\20+ yard plays\pbp-2020.csv'
df = pd.read_csv(pbp)
pd.set_option('display.max_columns', None)
df.head()
```

```python
notExplosive = df[df['Yards'] < 20].index
df.drop(notExplosive, inplace=True)
run = df[df['IsRush'] == 1].index
df.drop(run, inplace =True)
df.head()
```

```python
t=df[['OffenseTeam','Yards']]
t = t.groupby(['OffenseTeam']).sum()
t.head(32)
```

```python
t = t.sort_values(by='Yards', ascending=False)
t['Yards'].plot.bar()
print("20+ Yard Passing Plays By Team")
```
